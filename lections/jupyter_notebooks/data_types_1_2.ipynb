# String или строки в Python


## Строки как тип данных. Базовые операции
Что такое строка? По своей сути - тоже самое, что мы подразумеваем в обычной жизни. Строка это набор символов, будь то буквы или цифры. Главное помнить, что цифра в строке - это именно строка, а не цифра. Строка в python обозначается ковычками, двойными или одинарными,  не имеет значения. Самое главное везде использовать один и тот же тип ковычек. Создадим переменную s и запишем туда строку

```python
s = 'lesson 1'
s
```

Вот собственно так и выглядит строка и таким простым образом она создается. Почему переменную назвали не str, а s ? Потому что в Python str уже зарезервировано внутренними методами самого Python. Точно также нельзя создавать переменную int, float и другие названия, которые являются внутренними методами самого Python.
Также при работе со строками, да и в целом с любыми типами данных, важно помнить, что даже если у вас в строке будет 1 цифра, то от этого она не станет типом данных integer или float, она будет именно строкой. 

### Преобразование строк и чисел

```python
# Посмотрим на типы данных переменных s и переменной n
s = '1'
n = 1
print(type(s))
print(type(n))
```

Проверив тип данных, можем убедиться, что у нас s - это именно строка, а n - это integer. Поэтому важно это всегда учитывать. Можно ли преобразовать число, записанное в строке в тип данных integer или float? Да, можно. Python - это язык динамической типизации, и за счет этого он определяет сам типы данных и поэтому это может получиться. Точно также, можно преобразовать число в строку, да и в целом, при наличии желания, можно любой тип данных преобразовать в строку и иногда это действительно может потребоваться. Рассмотрим пока что пример с преобразованием строки в число и обратно в строку.

```python
# Преобразуем число 1 записанное в строке, в тип данных integer и float
s = '1'
print('type of s = ', type(s))
int_s = int(s)
print('type of int_s = ', type(int_s))
float_s = float(s)
print('type of float_s = ', type(float_s))

#По аналогии с числовыми типами данных используем str() для преобразования в строку

back_to_string_int = str(int_s)
print('type of back_to_string_int = ', type(back_to_string_int)) 
back_to_string_float = str(float_s)
print('type of back_to_string_float = ', type(back_to_string_float))
```

После преобразований такого рода с новыми переменными уже можно работать именно как с числами, а не как со строками. Что это значит? Сложим между собой строку переменную s в которой записана единица как строка и новые переменные int_s и float_s

```python
# Результат сложения двух строк
new_s = s + s
new_s
```

```python
# Результат сложения int_s и float_s переменных
new_s_numbers = int_s + float_s
new_s_numbers
```

Как видно, когда сложили 2 раза строки s, в результате, получили строку new_s, которая была равна не 2, а '11'. Т.к. когда происходит сложение строк, к первой строке, всегда с правой строны, добавляется вторая строка. Так можно складывать хоть 3 хот 15 строк, количество не важно. Но всегда прибавлением будет производиться к предшествующей строке с правой стороны. Чтобы было понятнее, посмотрим это на примере 3 строк

```python
#Сложение 3 строк 
s1 = 'строка номер 1'
s2 = 'новая строка номер 2'
s3 = 'последняя строка номер 3'

result_s = s1 + s2 + s3
result_s
```

### Измененяемость строк
Как это видно и из других примеров, каждый раз, чтобы чтобы преобразовать строку, создается новая переменная. Почему так? Вспомним к каким типам данных по изменяемости относится строка


Из рисунка видно, что строки являются неизменяемыми типами данных. Что это значит? Что если у нас есть строка s1 и мы будем ее просто менять не перезаписывая, к примеру, прибавляя к ней строку s2, то строка не изменится. Что это значит? Что если мы просто напишем s1 + s2 и выведем s1, то получим исходную строку s1. Если мы хотим все таки прибавить s2 к 1 s1, то надо либо перезаписать переменную s1 в которой у нас записана строка, либо создать новую переменную, как это делалось раньше. Рассмотрим на примерах в коде

```python
# Сложим строки не изменяя исходную и не создавая новую переменную
s1 + s2
s1
```

```python
# Теперь попробуем перезаписать уже существующую переменную s1
s1 = s1 + s2
s1
```

Как видно из примеров, если мы не перезаписываем, т.е. по сути не создаем новую строку, которая состоит уже изсуммы s1 и s2, то никакого эффекта не будет происходить. А теперь посмотрим на выражение s1 = s1 + s2, не похоже ли оно на выражение с числами, по типу x = x + 2 ? Похоже и даже очень. А можем ли мы сократить эту строку до записи как с числами x += 2 ? Да, можем, точно также записав s1 += s2.

```python
# Сокращенное сложение строк. Т.к. на предыдущем шаге мы уже перезаписали переменную s1, снова перезапишем ее в исходный вид
s1 = 'строка номер 1'
s1 += s2
s1
```

Как видно, строка тоже перезаписалась и объеденилась со строкой s2.

Также можно заметить, что строки объеденились но в тоже время, они при этом слиты и не разделены между собой. Нормально ли это? Да. Строки всегда будут объеденяться без какого-либо разделителя. Это можно поправить просто добавив строку содержащую нужный нам разделитель. Для этого не обязательно создавать отдельную переменную, можно просто создать строку в коде добавления. Разделять строки можно по разному, с помозью запятой, пробела, табуляции, точки запятой. Собственно от типа разделения строк в дальнейшем будут зависить и форматы файлов, к примеру, известный всем tsv - это tab separated, т.е. строки в нем разделены табуляцией, csv - comma separated - т.е. разделитель это запятая. Попробуем сделать строку с разделение с помощью табуляции и строку с разделение с помощью запятой

```python
# Разделим строку с помощью пробела, таблуляции и запятой. Также еще раз создадим переменные s1,s2,s3. 

s1 = 'строка номер 1'
s2 = 'новая строка номер 2'
s3 = 'последняя строка номер 3'

space_s = s1 + ' ' + s2 + ' ' + s3
tsv_s = s1 + '\t' + s2 + '\t' + s3
csv_s = s1 + ',' + s2 + ',' + s3

print(tsv_s)
print(csv_s)
```

Тут можно заметить конструкцию \t в строке, где применялось разделение табуляцией. Собственно символ \t - и означает табуляцию, а слэш перед ним, это то, что называется экранированием символа, т.е. чтобы код будет считываться код, можно было понять, что это именно символ табуляции, а не часть строки. Кроме символа табуляции, есть еще несколько наиболее распространненых, таких как к примеру перенос строки \n или перенос курсора(карретки) \r. В таблице ниже преведены ряд таких используемых символов

### Строки и индексы
Теперь понятно, какие базовые операции можно делать со строками. Что еще с ними можно делать, кроме базового их создания сложения? Вспомним, что кроме неизменяемости у строк есть еще одно свойство - упорядоченность. Что это значит? Что каждый символ в строке имеет четкую позицию, т.е. имеет индекс по которому к нему можно обратиться. Выглядит это так, допустим у нас есть строка s5 = 'alpacas'. Каждая буква тут имеет свой индекс, начиная от 0 и заканчивая последним символом. Это важный момент, что отсчет в Python начинается именно с 0, а не с 1! Выглядит это так:



Что это значит и что такое обратная позиция? И почему она начинается не с 0 ? Итак, позиция, это то, где для программы у вас в данной строке стоит каждый элемент. Это значит, что обратившися к определенному индексу, т.е. взяв лишь только элемент в какой-то позиции в строке, мы можем получить определенный сивол в этой позиции. Проще говоря, в строке alpacas, буква p стоит во 2 позиции, если считать от начала слова, т.к. отсчет идет от 0. Если мы обратимся к элементу строки, а обращение к объекту в Python делается через квадратные скобочки, то можем получить букву p. Т.к. позиция 2, значит индекс 2, обратимся ко второму элементу строки вот так s5[2]. И в таком случае полуим букву p. Что такое обратная позиция? Это когда, к примеру, нам надо взять последнюю букву, и чтобы не считать длину слова, можно просто взять последний элемент, написав s5[-1] и получим букву s. Посмотрим на это все в коде:

```python
# Получим букву p из строки alpacas
s5 = 'alpacas'
s5[2]
```

```python
# Получим последний символ строки с использованием прямого индекса (т.е. отсчет от начала слова)
s5[6]
```

```python
# Получим последний символ строки с использованием обратного индекса
s5[-1]
```

А как определить индекс элемента? Особенно как узнать длину строки? Легко, в Python существует метод len, который покажет "длинну" интересующего объекта. Т.е. по сути, он покажет вам сколько элементов присутсвует в объекте. Рассмотрим его применение на нашей строке s5

```python
len(s5)
```

Важно всегда помнить, что метод len считает от 1, а индексирование в Python начинает с 0. Т.е. в строке alpacas действительно 7 символов, но индекс последнего будет 6, а не 7.

### Срезы строк

Возникает вопрос, а можно ли как-то взять не одну букву, а несколько сразу? Ответ - да, можно, но только стоящие подряд или с каким-то шагом. Для этого используется метод срезов, широко используемый способ для многих типов данных метод. Выглядит он просто: мы также обращаемся к интересующей нас строке, а после этого просто напросто указываем через двоеточие [начало:конец, шаг с которым брать элементы]. Выглядит это так, если, к примеру, нам надо взять из строки все символы кроме 1, то просто напишем 
s5[1:]. Такая запись обозначает, что мы указываем только начало откуда брать срез, а конец оставляем поумолчанию, т.е. это конец нашей строки. Если нуны лишь элементы со 2 по 4, то выглядит это так s5[1:3]. Тут важно помнить 1 маленький ньюанс, с точки зрения математики такой срез бы записывался как [1:3), т.е. 4 элемент не включается, а значит чтобы действительно получить срез 3 элементов со 2 по 4 включительно, срез надо брать по 5 элемент, т.е. s5[1:4]. Также можно взять все элементы со 2 по предпоследний и с помощью прямой и обратной индексации одновременно, выглядит это вот так s5[1:-1]. Что же такое шаг? Это то, как будет считаться добавляемые символы. Т.е. отсчет при срезе начинается, к примеру, с 0 элемента, затем с шагом 1, прибавляет 1, и берется 1 элемент, затем прибавляется еще раз 1 и берется второй элемент и так далее. Если начало идет с другого элемента принцип остается тот же. Т.е. всегда когда вы не указываете шаг при срезе он равен 1. А если сделать шаг равный 2, то он возьмет каждый второй элемент от указанного начала, до указанного конца. Возьмем каждый 2 элемент в строке s5. Для этого нужно обязательно указать, что начало и конец это начало строки и конец строки и только после этого указать шаг! Выглядит это следующим образом:
s5[::2]. Что обозначает эта запись? Первое двоеточие без указаний на начало и конец, говорит нам о том, что начало и конец не изменяются, а второе двоеточее говорит, что для данного среза будет задаваться шаг, после которого уже пишется значение шага.
Посмотрим это все на примерах в коде

```python
#Срез строки s5 от 2 до последнего элемента
s5[1:]
```

```python
#Срез строки s5 от 2 до 4 элементо
s5[1:4]
```

```python
#Срез строки s5 от 2 до -1 элементо
s5[1:-1]
```

```python
#Использование шага 2 при взятии среза
s5[::2]
```

```python
#Использование шага 2 при взятии среза не от начала, а от 2 элемента
s5[1::2]
```

```python
#Использование среза и шага -1 для получение строки в обратном порядке
s5[::-1]
```

## Строки и методы строк
У всех типов данных начиная со строк, есть ряд встроенных методов, которые позволяют различным образом модифицировать эти типы данных. Разберем ряд наиболее популярных методов которые используются при работе со строками. Чтобы добавить щепотку биологии и уже смело начать называть себя начинающим бионинформатиком, возьмем начало нуклеотидной последовательности гена кодирующего natriuretic peptide A у альпака: GCTGGGGGGCTATAAAAAGAGGAGGCGCAGCGGAGAGACAGAGACGGACAAAGACCAAGAGAGCGAAACCAGAGAGGAGCAAGAAGCAGCAAACACCAGACCATTCTTTGACCGACTCCAGCATGGGCTCCTCCGCCATC
Как можно заметить, получится очень длинная строка. Согласно правилам хорошего года в Python строка не должна быть длиннее 79 символов. Проще говоря, если строка не влазит нормально в область зрения, то ее надо перенести. Для этого используется сивол \ , который говорит программе о том, что тут идет перенос кода на следующую строку.

```python
# Создадим переменную в которой будет лежать нуклеотидная последовательность
nppa_gene_start = ' GCTGGGGGGCTATAAAAAGAGGAGGCGCAGCGGAGAGACAGAGACGGACAAAGACCAAGAGAGCGAAACC \
AGAGAGGAGCAAGAAGCAGCAAACACCAGACCATTCTTTGACCGACTCCAGCATGGGCTCCTCCGCCATC '
```

Перейдем к самой работе со строкой. Как использовать методы, которые есть у типа данных строка? Также, как использовать методы у библиотек. 'строка'.'интересующий меотд'(). В скобочках могут указываться дополнительные условия, которые есть у метода.Начнем с самого популярного и важного метода при работе со строками, особенно полученными из файлов - strip. Это метод, который преобразует строку убирая пробельные символы перед и в конце строки. Важно понимать, что пробельный символ и табуляция, это 2 разных символа и табуляцию метод strip не убирает. Если нет необходимости убирать с обеих сторон пробельные символы, а только слева или справа, есть методы lstrip и rstrip соответсвенно. Если применить обычный strip к строке, где слева есть пробельный символ, а справа его, к примеру, нет, то метод просто уберет символ слева без каких-либо проблем.
Как видно, в переменной nppa_gene_start умышленно добавлены пробелы перед и в конце строки. Удалим пробельные символы. Тут стоит отметить, что ЛЮБЫЕ методы строк - изменяют строку, а это значит, что ВСЕГДА нужно перезаписывать переменную или создавать новую, а иначе изменения не произойдут.

```python
nppa_gene_start = nppa_gene_start.strip()
nppa_gene_start
```

Как можно заметить, концевые пробельные символы удалились. Теперь при внимательном просмотре переменной можно увидеть, что кажется, пробельный символ закрался в середину строки. А метод strip удаляет лишь только концевые пробельные символы. Решение этого уже зависит от природы этого пробела. Допустим, это опечатка при внесении данных, тогда надо удалить этот пробел. Для этого существует метод replace(что, на что, максимальное количество замен) с 3 параметрами. Первые 2 - обязательные, 3 - не обязательный. метод replace может не только удалить пробелы, но и, к примеру, заменить одни символы на другие. Получается, что чтобы убрать пробел, нам надо заменить все пробельные символы, на ничего, выглядит это так replace(' ', ''). Посмотрим как это работает в коде

```python
# Удалим пробел из строки с помощью replace
nppa_gene_start_nosp = nppa_gene_start.replace(' ', '')
nppa_gene_start_nosp
```

Теперь пробельного символа нет и строка выглядит так, как должна выглядеть. Также с помозью replace, можно заменять любые сиволы.
Используемая последовательность - это последовательность ДНК, это можно понять по наличию нуклеотида T - тимина, т.к. в последовательности РНК вместо тимина присутсвует урасил с символом U. Попробуем превратить нашу ДНК последовательность в РНК, путем замены всех тиминов на урацилы

```python
# Заменим для начала только 1 тимин
rna_nppa_gene_start_nosp = nppa_gene_start_nosp.replace('T','U', 1)
rna_nppa_gene_start_nosp
```

```python
# А теперь заменим все оставшиеся тимины
rna_nppa_gene_start_nosp = rna_nppa_gene_start_nosp.replace('T', 'U')
rna_nppa_gene_start_nosp
```

Теперь рассмотрим ситуацию, что тот пробельный символ это не опечатка - а это авторское решение так разделить 2 последовательности. Т.е. теперь это значит что у нас в одной строке записаны 2 последовательности и нам может потребоваться работать по отдельности с каждой. Более того, пробельный символ не самый надежный разделитель, лучше использовать либо символ табуляции, либо точку с запятой. Или можно использовать перенос строки, чтобы каждая последовательность была на своей строчке. Но как это сделать? Чтобы просто разделить строки правильным разделителем, можно снова воспользоваться методом replace, а вот чтобы отделить эту строку на 2 можно (и нужно ) воспользоваться меотдом split('Нужный разделитель'). Метод split имеет лишь 1 аргумент - разделитель,по которому будет делиться исходная строка. Указанный разделитель после этого удаляется и строки преобразуются в список, тип данных про который будет расказано чуть позже. В этом списке каждая строка будет уже лежать отдельно и можно будет отдельно работать с каждой.  
Важно помнить, что метод split удаляет разделитель при делении строк!

```python
#Разделим исходную строку на 2 подстроки по пробелу в качестве разделителя
nppa_gene_start_split = nppa_gene_start.split(' ')
nppa_gene_start_split
```

В результате исходная строка разделилась на 2, пробельный символ удалился и можно работать с каждой строкой поотдельности. Как это делать будет объяснено в разделе про типы данных списки. Кроме каких-то стандартных разделителей, строку можно разбить вообзе по любому разделителю, даже по 1 букве или по набору букв. Разобъем исходную последовательность по подпоследовательностям по посторяющимся нуклеотидам цитозина, т.е. 'CC'. Таким образом получим все участки последовательности между которыми был именно такой набор нуклеотидов. Но в нашей исходной последовательности, по прежнему имеется пробельный символ, т.е. его тоже надо удалить. Можно взять уже последовательность с удаленным пробельным символом, но предположим, что такой еще нет. В итоге, код должен выглядеть следующим образом:

nppa_gene_start = nppa_gene_start.replace(' ', '')  
nppa_gene_start_without_cc = nppa_gene_start.split('CC')

И тут все правильно, но в Python есть определенное правило написание кода, что если делается несколько модификаций 1 объекта подряд, лучше стоит написать их в одной строке. Так код будет проще читать и понимать. Но в целом, писать каждый шаг по отдельности не запрещается. Тут данный пример будет преведен для того, чтобы показать как в Python можно последовательно использовать нескольно методов. Вылглядеть код будет так:

```python
nppa_gene_start_without_cc = nppa_gene_start.replace(' ', '').split('CC')
nppa_gene_start_without_cc
```

Как итог, 1 строкой мы сначала удалили пробельный символ, а затем разбили на подстроки. Самое главное не забывать, что тут очень важна последовательность действий! Т.к. если бы строка была записана таким образом:  
nppa_gene_start_without_cc = nppa_gene_start.split('CC').replace(' ', '')  
То ничего бы не получилось и код выдал бы ошибку. Почему? Потому что  действия выполняются последовательно, а значит, сначала метод split преобразует строку к массиву, а затем код пытается применить метод типа данных строка replace к типу данных список, а у список нет метода replace.   

Итак, в результате получили 11 отдельных строк и удалили все нуклиотидные последовательности 'CC' из исходной. Но, предположим, что исходную последовательность забыли сохранить и она пропала в пространстве и времени. А нужна снова исходная последовательность (ну только без пробельного символа). Что делать в таком случае? Тут на помощь приходит метод join, который позволяет создать строку из списка с указанным разделителем. Т.е. по сути, join это просто метод split наоборот. Работает он следующим образом: 'строка по которой объеденить объекты списка'.join('список который надо объеденить'). Попробуем востановить исходную строку, получается необходимо полученный после разделения по 'CC' список преобразовать в строку с разделителем строки 'CC'. Выглядеть это будет вот так:

ДОБАВИТЬ ПРО ВАРИАНТ JOIN 2

```python
#использование join вариантом №1
return_to_string = 'CC'.join(nppa_gene_start_without_cc)
return_to_string
```

Таким образом можно подсчитывать любые символы в строке и определять, к примеру, какой тип нуклеотидов или какой нуклеотид доминирует в последовательности. А можно ли найти какую-то конкретную подстроку? Да, но к сожалению, стандартные методы строки позволяют найти лишь 1 или последнее совпадение. Позиция - это где элемент начинается, т.е. если искомая подстрока длиннее 1 элемента, то find выдаст индекс где данная подстрока начинается. Если поиск не дал результатов, то find или rfind напишет -1. Также в Python есть аналог find и rfind - index. Но если вы не уверены в наличии подстроки рекомендуется использовать именно find, т.к. ValueError может появиться не только из-за отсутсвия подстроки.

## Условные операторы. Конструкции if, else, elif

Еще одной важно вещью являются условные операторы. Это такие конструкции, которые, как это ни странно, проверяют заданные условия
базовая конструкция выглядит так:  
if 'условие':  
    'какое-от действие'  
else:  
    'какое-то действие'  
Проще говоря это выглядит так: если условие выполняется, то сделать действие 1, иначе, сделать действие 2.

Рассмотрим простой код оценивающий длинну строки

```python
# Задача для курса по раздзелу строки. Может быть дать это в домашку?
```

```python
# оценим длину строки и если она больше 10, то напечатаем ее, а если меньше 10, то добавим к ней еще одну строку из _,
# чтобы длина стала больше 10
s9 = input()
if len(s9) > 10:
    print('Данная строка длинее 10 символов', s9, sep='\n')
else:
    new_s9 = s9 + ((10-len(s9)) * '_')
    print('Строка была короче 10 символов на {}, пришлось модифицировать'.format(10 - len(s9)),
          'Полученная строка:', new_s9, sep='\n')
```

Что сделали в примере выше?
1) Использовали функцию input, чтобы не создавать строку, а вводить ее из консоли и записали в переменную s9
2) Начали конструкцию if и задали условие: Если длинна полученной строки s9 больше 10, то вызвать функцию print() и напечатать ей переданную строку и описание.
3) Продолжили логическую конструкцию с помощью else, где говорится, что если условие переданное в if не выполняется, т.е. длинна строки меньше 10, в таком случае, проверяем с помощью оператора len сколько не хватает до длинны в 10 и с помощью оператора * создаем строку из символов _ такой длины, которой не хватает до 10 символов и добавляем ее к строке s9 и записываем в новую переменную new_s9. 
4) затем вызываем оператор print() и печатаем new_s9 с заданными параметрами

На первый взгляд может показаться, что в целом это не сильно применимая задача. Но если предположить, что вы работаете с биоинформатической программой, которая не принимает последовательности меньше 10 нуклеотидов? Тогда вместо символа нижнего подчеркивания поставим символ 'N' который обозначает неизвестно какой нуклеотид в данной позиции и посмотрим на этот код еще раз


```python
,nuc_line = input()
if len(nuc_line) > 10:
    print('Нуклеотидная последовательность длинее 10 символов', nuc_line, sep='\n')
else:
    new_nuc = nuc_line + ((10-len(nuc_line)) * 'N')
    print('Нуклеотидная последовательность короче на {} символов, пришлось модифицировать'.format(10 - len(nuc_line)), 
          'Модифицированная последовательность:', new_nuc, sep='\n')
```

Самое главное при использовании конструкции if else помнить, что она ОБЯЗАТЕЛЬНО должна содержать оба компанента, т.е. должно быть условие if и должно быть else когда условие в if не выполняется. Кроме базовой конструкции if else, есть еще оодна, которая на самом деле применяется даже чаще, конструкция if elif else. В этой конструкции задается изначальное условие, а затем идут дополнительные условия. elif может быть сколько угодно и самое главное, в этой конструкции наличие else не обязательно. Но тут стоит помнить, что условия проверяются подряд и если условие в if выполнено, то проверка в elif уже не пойдет, как это продемонстрировано в примере ниже, где несмотря на то, что a == 2, 2 > 0 и программа не проверяет уже условия после if

```python
a = 2
if a > 0:
    print(1)
elif a == 2:
    print(2)
```

В общем виде работа if elif else выглядит так:

if 'условие':
'какое-от действие'
elif 'еще условие, если условие в if не выполнено':
'какое-от действие'
elif 'еще условие, если условие в предыдущем elif не выполнено':
'какое-от действие'
else:
'какое-то действие'
Проще говоря это выглядит так: если условие 1 выполняется, то сделать действие 1, если условие один не выполняется, проверить условие 2 и выполнить действие 2, если условие 2 не выполняется, проверить условие 3 и выполнить действие 3, иначе, т.е. если все условия не выполняются, сделать действие 4.
