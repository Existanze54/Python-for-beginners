# Задача 1

- сгененрируйте 2-х мерный список с числами
- постройте для этого списка обычный линейный график
- постройте для этого списка обычный точечный график
- подпишите оси и сами графики

```python
# Место для решения задачи
import matplotlib.pyplot as plt

import numpy as np
list_of_numbers_first = [np.random.randint(1, 20, size = (10)), np.random.randint(1, 20, size = (10))]
print(list_of_numbers_first)

plt.plot(list_of_numbers_first[0], list_of_numbers_first[1])
plt.plot(list_of_numbers_first[0], list_of_numbers_first[1], 'ro')
plt.title('Random graph')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

# Задача 2

- Сгенерируйте второй 2 мерный список, где значения оси y это логарифмированные значения оси y из первого списка по основанию 10.
- Постройте график для обоих списков и подпишите их
- Постройте график для первого списка и вложенный на нем график для второго списка. При этом ось x у вложенного списка теперь должна быть логарифмированные значения по натуральному основанию

```python
# Место для решения задачи
import math
import numpy as np
#list_of_numbers_first = [np.random.randint(1, 20, size = (10)), np.random.randint(1, 20, size = (10))]
list_of_numbers_second = [list_of_numbers_first[0], [math.log10(x) for x in list_of_numbers_first[0]]]
#print(list_of_numbers_second)


fig = plt.figure(figsize=(10, 6))
axes = fig.add_axes([1,1,1,1])
axes.plot(list_of_numbers_first[0], list_of_numbers_first[1], 'b')
axes.set_xlabel('x') 
axes.set_ylabel('y')
axes.set_title('First graph') 
axes.legend(['random (x,y)'], loc='best')
plt.show()

fig = plt.figure(figsize=(10, 6))
axes = fig.add_axes([1,1,1,1])
axes.plot(list_of_numbers_second[0], list_of_numbers_second[1], 'g')
axes.set_xlabel('x') 
axes.set_ylabel('y')
axes.set_title('Second graph') 
axes.legend(['y=log10(x)'], loc='best')
plt.show()



list_of_numbers_third = [list_of_numbers_first[0], [math.log(x) for x in list_of_numbers_first[0]]]

fig = plt.figure(figsize=(10, 6))
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # основной график
axes2 = fig.add_axes([0.46, 0.25, 0.4, 0.2]) # внутренний график
# Основной график
axes1.plot(list_of_numbers_second[0], list_of_numbers_second[1], 'b')
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes1.set_title('Внешний график')
axes1.legend(['y=log10(x)'], loc='best')
# Вложенный график
axes2.plot(list_of_numbers_third[0], list_of_numbers_third[1], 'g')
axes2.set_xlabel('x1')
axes2.set_ylabel('y1')
axes2.set_title('Внутренний график')
axes2.legend(['y=log(x)'], loc='best')
plt.show()
```

# Задача 3

Постройте subplot размером 2-2 (т.е. 4 графика). На первом subplot отобразите график в котором сложите оси y из 1 и в торого списка, у второго разность, у третьего произведение и у четвертого деление. Подпишите графики

```python
# Место для решения задачи

plt.figure(figsize=(16, 8))

plt.subplot(2,2,1)
plt.plot(list_of_numbers_first[0], list_of_numbers_first[1] + list_of_numbers_second[1])
plt.title('Graph 1')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(['y1+y2'], loc='best')

plt.subplot(2,2,2)
plt.plot(list_of_numbers_first[0], list_of_numbers_second[1] - list_of_numbers_first[1])
plt.title('Graph 2')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(['y2-y1'], loc='best')

plt.subplot(2,2,3)
plt.plot(list_of_numbers_first[0], list_of_numbers_first[1] * list_of_numbers_second[1])
plt.title('Graph 3')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(['y1*y2'], loc='best')

plt.subplot(2,2,4)
plt.plot(list_of_numbers_first[0], list_of_numbers_second[1] / list_of_numbers_first[1])
plt.title('Graph 3')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(['y2/y1'], loc='best')

# Зададим растояние между графиками
plt.subplots_adjust(wspace=0.2, hspace = 0.35)

plt.show()
```

# Задача 4

- для полученной таблицы из файла vac_bsg.txt, постройте график, чтобы визуально сравнить количетво рожденных детей и количество детей, которых должны были вакцинировать
- постройте sublot из 3 графиков, где отобразите сравнение медианы и среднего для вакцинированных детей с помощью графиков bar plot и scatter plot, а также постройте гистограмму рожденных детей. Подпишите все оси, графики и легенды

Для задачи можно использовать как Matplotlib, так и Seaborn

```python
# Место для решения задачи
import os
path_to_code = os.getcwd()
path_to_inp_file = os.path.join(path_to_code, os.pardir, 'lections', 'files_and_images', 'vac_bsg.txt')

import matplotlib.pyplot as plt

import seaborn as sns
sns.set(rc={'figure.figsize':(16, 8)})
sns.set_style("darkgrid")

import pandas as pd
vac_bsg = pd.read_table(path_to_inp_file, sep = '\t', encoding='cp1252')

font1 = {'family': 'Georgia', # это шрифт
        'color':  'green', # Цвет шрифта
        'weight': 'normal', # тип шрифта (жирный, полужирный и т.п.)
        'size': 16, # Размер Шрифта
        }

ch_births = vac_bsg['BirthsUNPD']
ch_in_targ = vac_bsg['ChildrenInTarget']
names = ['Births', 'In target']
values = [ch_births.sum(), ch_in_targ.sum()]

plt.figure(figsize=(16, 8))
plt.bar(names, values, color='g')
plt.title('Total number of children born/to be vaccinated since 1997', fontdict = font1)
plt.show()

font2 = {'family': 'Georgia', # это шрифт
        'color':  'blue', # Цвет шрифта
        'weight': 'normal', # тип шрифта (жирный, полужирный и т.п.)
        'size': 11, # Размер Шрифта
        }

ch_vac = vac_bsg['ChildrenVaccinated']
names_for_subplot = ['Median', 'Mean']
values_for_subplot = [ch_vac.median(), ch_vac.mean()]

plt.figure(figsize=(16, 8))
plt.subplot(1,3,1)
plt.bar(names_for_subplot, values_for_subplot, color='b')
plt.title('Median and mean value of vaccinated children since 1997', fontdict = font2)

plt.subplot(1,3,2)
plt.scatter(names_for_subplot, values_for_subplot, color='r')
plt.title('Median and mean value of vaccinated children since 1997', fontdict = font2)

plt.subplot(1,3,3)
plt.hist(vac_bsg['Year'])
plt.title('Histogram of children born since 1997', fontdict = font2)
plt.show()
```
