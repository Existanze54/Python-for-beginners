```python
# Место для решения
import os
path_to_code = os.getcwd()
path_to_inp_file = os.path.join(path_to_code, os.pardir, 'lections', 'files_and_images', 'vac_bsg.txt')
#print(path_to_code)
#print(path_to_file)

with open(path_to_inp_file, encoding = 'cp1252', mode = 'r') as inp_file:
    vac_bsg_original = pd.read_csv(inp_file, sep = '\t')
vac_bsg = vac_bsg_original.copy()

vac_bsg_sorted = vac_bsg[(vac_bsg['Vaccine'] == 'hepb3') & (vac_bsg['Country'] == 'Russian Federation')] 
vac_bsg_sorted
```

# Задача 4

- Удалите столбец Comment
- Создайте столбец в котром будет содержаться информация о разнице между ChildrenInTarget и ChildrenVaccinated

```python
# Место для решения
vac_bsg['ISOCountryCode'] = vac_bsg['ISOCountryCode'].apply(lambda x : x.upper())
vac_bsg.drop('Comment', axis = 1, inplace = True)
vac_bsg.insert(12, 'ChildrenDiff', vac_bsg['ChildrenInTarget'] - vac_bsg['ChildrenVaccinated'])
vac_bsg
```
